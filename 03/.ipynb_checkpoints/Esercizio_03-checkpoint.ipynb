Ho calcolato il valore di opzioni call e put usando due metodi diversi: un campionamento diretto e un campionamento discreto. 

Per il campionamento diretto ho calcolato il prezzo $S(T)$ di un'azione noto il prezzo attuale $S(0)$ tramite la formula: 
$$S(T)= S(0) \cdot \exp{\big[(r-\frac{\sigma^2}{2})T + \sigma z \sqrt{T}\big]} ,$$
dove $z$ è generato casualmente tramite una distribuzione gaussiana con $\mu=0$ e $\sigma=1$, mentre $r$ e $\sigma$ sono rispettivamente il tasso di interesse e la volatilità del mercato. Da $S(T)$ è possibile stimare il prezzo $C$ delle opzioni call e il prezzo $P$ delle opzioni put tramite le relazioni:
$$C = e^{-rT} \cdot \max{\big(0,S(T)-K\big)}$$
$$P = e^{-rT} \cdot \max{\big(0,K-S(T)\big)}$$
dove $K$ è il prezzo pattuito allo scadere dell'opzione. Quindi, ho calcolato $10^6$ volte $C$ (o $P$) trovandone media aritmetica ed incertezza statistica. Per velocizzare il calcolo, ho diviso in 100 blocchi i risultati, e ho riportato i prezzi all'aumentare del numero di blocchi presi in considerazione.

Per il campionamento discreto invece il calcolo di $S(T)$ è stato condotto dividendo in 100 sezioni uguali l'intervallo di tempo $(0,T)$, calcolando $S(t_{i+1})$ s partire da $S(t_i)$ tramite la relazione: 
$$S(t_{i+1})= S(t_i) \cdot \exp{\big[(r-\frac{\sigma^2}{2})\Delta t + \sigma z \sqrt{\Delta t}\big]} ,$$
dove $\Delta t = t_{i+1} - t_i $ e $z$ è generato in modo analogo al caso del campionamento diretto. Ponendo $S(t_0)=S(0)$ e $S(t_{100})=S(T)$ si arriva percorrendo questi 100 passi alla stima di $S(T)$. Da qui in poi si procede in modo analogo al caso precedente.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from astropy.table import Table

Names = np.array(["dir_mean","dir_errs","paths_mean","paths_errs"])
datas = Table.read("Results.txt",format="ascii",names=Names)

dir_mean = np.array(datas["dir_mean"])
dir_errs = np.array(datas["dir_errs"])
paths_mean = np.array(datas["paths_mean"])
paths_err = np.array(datas["paths_errs"])
N = len(dir_mean)
x = np.arange(N)
    
plt.title("Opzione call con sample diretto")
plt.errorbar(x,dir_mean,yerr=dir_errs)
plt.axhline(14.975790778311286,color="r") #valore preso dal valore teorico presente nella consegna dell'esercizio
plt.xlabel("# blocchi considerati")
plt.ylabel("Valore dell'opzione")
plt.ylim(14.85,15.15)
plt.grid(True)
plt.show()

plt.title("Opzione call con sample discreto")
plt.errorbar(x,paths_mean,yerr=paths_err)
plt.axhline(14.975790778311286,color="r")
plt.xlabel("# blocchi considerati")
plt.ylabel("Valore dell'opzione")
plt.ylim(14.85,15.15)
plt.grid(True)
plt.show()
```

Si nota che entrambi i metodi danno incertezze simili e entrambi i risultati sono a meno di un $\sigma$ dal valore teorico. Ho quindi ottenuto risultati validi.

```python
Names = np.array(["dir_mean","dir_errs","paths_mean","paths_errs"])
datas = Table.read("Results_put.txt",format="ascii",names=Names)

dir_mean = np.array(datas["dir_mean"])
dir_errs = np.array(datas["dir_errs"])
paths_mean = np.array(datas["paths_mean"])
paths_err = np.array(datas["paths_errs"])
N = len(dir_mean)
x = np.arange(N)
    
plt.title("Opzione put con sample diretto")
plt.errorbar(x,dir_mean,yerr=dir_errs)
plt.xlabel("# blocchi considerati")
plt.ylabel("Valore dell'opzione")
plt.ylim(5.35,5.55)
plt.axhline(5.4595325819072364,color="r") #valore uguale al valore teorico sulla consegna dell'esercizio
plt.grid(True)
plt.show()

plt.title("Opzione put con sample discreto")
plt.xlabel("# blocchi considerati")
plt.ylabel("Valore dell'opzione")
plt.ylim(5.35,5.55)
plt.errorbar(x,paths_mean,yerr=paths_err)
plt.axhline(5.4595325819072364,color="r")
plt.grid(True)
plt.show()
```

Anche in questo secondo caso ho errori simili per entrambi i casi. Questa volta l'errore relativo al sample discreto è sopra un $\sigma$, ma essendo sotto $2\sigma$ lo ritengo comunque accettabile. 

Confrontando i quattro grafici, non ho notato differenze tali da poter stabilire se un metodo di campionamento dia risultati migliori dell'altro. L'unica chiara differenza è la miglior efficienza computazionale del campionamento diretto, in quanto non compie i 100 step per arrivare alla stima di $S(T)$ e dà direttamente il risultato finale.
