## <span style="color:green"> Esercizio 07.2 </span>

Per verificare l'andamento dell'autocorrelazione e della deviazione standard in funzione della lunghezza $L$ di un blocco ho introdotto il codice sottostante. I dati analizzati sono stati ottenuti da una simulazione fatta usando lunghezza dei blocchi $L=1$ e $5\cdot10^5$ blocchi. Ho usato l'energia potenziale e quella totale come dati, ma ho riportato solo quelli relativi all'energia potenziale in quanto in nessuno dei casi un cambio della  hanno dato risultati differenti.

Per il calcolo dell'autocorrelazione ho semplicemente usato il codice presente nella libreria statmodels (avevo provato un'implementazione diversa dell'algoritmo ma risultava troppo lento per essere pratico). Per il calcolo dell'errore medio ho preso $\forall L$ 500 blocchi casuali di lunghezza $L$ e ne ho mediato le deviazioni standard. Per questo motivo nel grafico della deviazione in funzione di $L$ mi aspetto una certa variazione casuale.



```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def read(filename) :
    file = open(filename,"r")
    value = []
    error = []

    for line in file:
        line = line.strip()
        lines = line.split()
        value.append(float(lines[2]))
        error.append(float(lines[3]))
        
    file.close()

    return value,error


import statsmodels.tsa.stattools

def autocorr_plots(nomefile) :

    pot_autocorr, trash = read(nomefile)

    acf=statsmodels.tsa.stattools.acf(pot_autocorr,nlags=1.5*10**5)
    N=np.linspace(0,1.5*10**5,num=len(acf))
    plt.plot(N,acf)
    plt.title("Autocorrelazione in funzione di $L \in [0,5000]$")
    plt.xlabel("$L$")
    plt.ylabel("Autocorrelazione")
    plt.xlim(0,5000)
    plt.show()
    plt.plot(N,acf)
    plt.title("Autocorrelazione in funzione di $L \in [0,1.5\cdot10^5]$")
    plt.xlabel("$L$")
    plt.ylabel("Autocorrelazione")
    plt.show()
    
def dev_stds_graph(nomefile) :
    pot_autocorr, trash = read(nomefile)
    dev_stds = np.zeros(100)
    print('-'*100)
    for k in range(0,100) :
        j = int(10+k*50000/100)
        #print(j,end='-')
        provv = np.zeros(500)
        print('-',end='')
        for i in range(0,500) :
            h = np.random.randint(0,len(pot_autocorr)-j-1)
            #print(h)
            provv[i]=np.std(pot_autocorr[h:(h+j)])
        dev_stds[k]=np.mean(provv)
    #print(sum)
    N = np.linspace(10,50000,num=len(dev_stds))
    xs=np.arange(len(dev_stds))
    xs=xs*50000/100
    plt.plot(xs,dev_stds)
    plt.title("Errore in funzione di $L$")
    plt.xlabel("$L$")
    plt.ylabel("Errore per blocco")
    plt.show()

def dev_stds_graph_bis(nomefile) :
    pot_autocorr, trash = read(nomefile)
    dev_stds = np.zeros(100)
    print('-'*100)
    for k in range(0,100) :
        j = int(10+k*5000/100)
        #print(j,end='-')
        provv = np.zeros(500)
        print('-',end='')
        for i in range(0,500) :
            h = np.random.randint(0,len(pot_autocorr)-j-1)
            #print(h)
            provv[i]=np.std(pot_autocorr[h:(h+j)])
        dev_stds[k]=np.mean(provv)
    #print(sum)
    N = np.linspace(10,5000,num=len(dev_stds))
    xs=np.arange(len(dev_stds))
    xs=xs*5000/100
    plt.plot(xs,dev_stds)
    plt.title("Errore in funzione di $L$")
    plt.xlabel("$L$")
    plt.ylabel("Errore per blocco")
    plt.show()
```

### <span style="color:blue"> Caso Solido ($T^* = 0.8$, $\rho^* = 1.1$, $r_{cut} = 2.2$  ) </span>

```python
#per il solido 0.1
autocorr_plots("Solid_Runs/Solid_corrected_1stRun/output_epot.dat")
dev_stds_graph("Solid_Runs/Solid_corrected_1stRun/output_epot.dat")
```

Vediamo in questo caso un'autocorrelazione che rimane alta per $L$ anche molto grandi. Notiamo ugualmente un errore continua ad aumentare anche per $L \approx 4\cdot10^4$, a significare che i passi rimangono correlati anche per lunghezze notevoli. 

Questo suggerisce che sono necessari blocchi molto grandi per evitare correlazioni, anche se difficilmente sarà praticabile sceglierne di dimensioni simili a quelle necessarie per annullarla completamente.

### <span style="color:blue"> Caso Liquido ($T^* = 1.1$, $\rho^* = 0.8$, $r_{cut} = 2.5$ ) </span>

```python
autocorr_plots('Liquid_runs/1st_Run/output_epot.dat')
dev_stds_graph_bis('Liquid_runs/1st_Run/output_epot.dat')
```

In questo caso invece possiamo vedere come l'autocorrelazione scenda a valori sufficientemente bassi già dopo $L\approx1000$, e come ad $L$ di questo tipo anche l'errore rimanga circa costante. Posso trarre quest'ultima conclusione vedendo che le oscillazioni statistiche sono dominanti rispetto ad un possibile aumento dell'errore con l'aumento di $L$.

Per questo nel caso liquido basterebbero blocchi di circa 1000 passi per evitare autocorrelazione indesiderata.

## <span style="color:green"> Esercizio 07.3 </span>

Ho aggiunto al codice il calcolo della quantità $\langle g(r) \rangle$, con $r \in [0,\frac{l}{2}]$, dove $l$ è il lato del cubo dove avviene la simulazione, tramite la formula:
$$\langle g(r) \rangle = \frac{2S}{\rho \Delta V N} ,$$
dove $\rho$ è la densità del materiale, $\Delta V = \frac{4 \pi}{3}\big( \frac{l}{2} \big)^3$ , $N$ il numero di particelle totali ed $S$ la somma delle particelle con una distanza reciproca minore di $\frac{l}{2}$.

## <span style="color:green"> Esercizio 07.4 </span>

Adesso ho calcolato tramite simulazioni di 20 blocchi di lunghezza $L=10^4$ l'energia potenziale, l'energia totale, la $\langle g(r) \rangle$ e la pressione del materiale in esame. Ho compiuto simulazioni sia usando il metodo Monte Carlo sia usando l'algoritmo di Verlet NVE, poi ne ho confrontato i risultati.

Mi aspetto che i risultati divergano almeno in parte a causa del fatto che l'algoritmo NVE non riesce ad ottenere esattamente la temperatura richiesta, ma risulta lievemente diversa per ogni simulazione. Questo fatto si ripercuote soprattutto sull'energia totale, in quanto una temperatura diversa implica un'energia cinetica diversa nei due casi, ma ha ovviamente effetto anche su tutte le altre variabili.

```python
def  final_graphs(nomecartella,oldrun) :
    
    value_press,error_press = read(nomecartella+"output_press.dat")
    value_g,error_g = read(nomecartella+"output_g.dat")
    value_etot,error_etot = read(nomecartella+"output_etot.dat")
    value_epot,error_epot = read(nomecartella+"output_epot.dat")
    
    oldvalue_press,olderror_press = read(oldrun+"output_press.dat")
    oldvalue_g,olderror_g = read(oldrun+"output_g.dat")
    oldvalue_etot,olderror_etot = read(oldrun+"output_etot.dat")
    oldvalue_epot,olderror_epot = read(oldrun+"output_epot.dat")
    
    value_temp, void = read(oldrun+"output_temp.dat")
    

    N=len(value_press)
    xs=np.arange(N)

    plt.errorbar(xs,value_press,yerr=error_press,label="MC")
    plt.errorbar(xs,oldvalue_press,yerr=olderror_press,label="NVE")
    plt.title("Pressione")
    plt.xlabel("# blocchi")
    plt.ylabel("Pressione")
    plt.legend()
    plt.show()
    plt.errorbar(xs,value_g,yerr=error_g,label="MC")
    plt.errorbar(xs,oldvalue_g,yerr=olderror_g,label="NVE")
    plt.title("$ < g(r) > $")
    plt.xlabel("# blocchi")
    plt.ylabel("$ < g(r) > $")
    plt.legend()
    plt.show()
    plt.errorbar(xs,value_epot,yerr=error_epot,label="MC")
    plt.errorbar(xs,oldvalue_epot,yerr=olderror_epot,label="NVE")
    plt.title("Energia Potenziale per particella")
    plt.ylabel("Energia Potenziale")
    plt.xlabel("# blocchi")
    plt.legend()
    plt.show()
    plt.errorbar(xs,value_etot,yerr=error_etot,label="MC")
    plt.errorbar(xs,oldvalue_etot,yerr=olderror_etot,label="NVE")
    plt.title("Energia Totale per particella")
    plt.ylabel("Energia Totale")
    plt.xlabel("# blocchi")
    plt.legend()
    plt.show()
    
    print("Temperatura media della simulazione NVE :" ,'%1.3f'%np.mean(value_temp))
```

### <span style="color:blue"> Caso Solido ($T^* = 0.8$, $\rho^* = 1.1$, $r_{cut} = 2.2$  ) </span>

```python
final_graphs("LAST_TRY/Solid_runs/2nd_run/","OLD_RUNS/Solid_runs/")
```

Notiamo che in questo caso le quantità ottenute non sono compatibili all'interno dell'errore dato. Questo si può sicuramente spiegare almeno in parte con la diffenerza di temperatura tra i due algoritmi, che risulta importante dati gli errori molto piccoli ottenuti (rispetto al valore totale delle grandezze). 
Notiamo inoltre che l'errore dell'algoritmo MC è più alto di quello NVE nella stima delle energie, più basso nella stima di $\langle g(r) \rangle$ e comparabile per la pressione.

### <span style="color:blue"> Caso Liquido ($T^* = 1.1$, $\rho^* = 0.8$, $r_{cut} = 2.5$ ) </span>

```python
final_graphs("LAST_TRY/Liquid_runs/2nd_run/","OLD_RUNS/Liquid_runs/")
```

Anche in questo caso otteniamo un risultato simile al caso solido. Coerentemente con le attese, l'energia totale è la quantità per la quale i due algoritmi risultano più discordanti.

### <span style="color:blue"> Caso Gassoso ($T^* = 1.2$, $\rho^* = 0.05$, $r_{cut} = 5.0$ ) </span>

```python
final_graphs("LAST_TRY/Gas_runs/2nd_run/","OLD_RUNS/Gas_runs/")
```

I risultati sono notevolmente più vicini nel caso gassoso. Infatti possiamo dire che i risultati sono compatibili ad un $\sigma$ per energia potenziale e $\langle g(r) \rangle$ e a 2$\sigma$ per la pressione. Per l'energia totale i valori rimangono notevolmente discordanti.

Per migliorare questi risultati ed ottenere valori più vicini tra le due simulazioni sarebbe necessario trovare un modo per far sì che la temperatura media dell'algoritmo NVE coincida più da vicino con quella impostata.
