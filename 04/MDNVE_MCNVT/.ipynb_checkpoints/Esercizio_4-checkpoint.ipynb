## <span style="color:green"> Esercizio 04.2 </span>

Ho quindi compiuto una simulazione di MD (Molecular Dynamics) per un solido, un liquido e un gas di Argon tramite un modello Lennard-Jones. Tuttavia, in tutti e tre i casi, è stato necessario fare simulazioni ulteriori per garantire la convergenza dell'algoritmo ai valori richiesti per $T$ e $\rho$. \
Infatti, all'inizio mi è stata data come situazione iniziale quella di un reticolo cristallino perfetto, ovviamente lontano da un caso realistico. Quindi ho fatto ogni volta una prima simulazione di setup per ottenere una disposizione degli atomi realistica e poi ho compiuto la simulazione vera e propria. \
La simulazione richiede in input le posizioni di partenza delle particelle, una temperatura $T^*$ e una densità $\rho^*$. Per fare la simulazione di setup è necessario impostare una temperatura $\mathcal{T}$ tale da ottenere alla fine atomi aventi la temperatura $T^*$ richiesta, ma in generale vale che $\mathcal{T} \neq T^*$. Per questo motivo ho compiuto più simulazioni di setup, in modo da stabilire un valore più adatto per $\mathcal{T}$. Ho riportato in questo notebook solamente i valori della simulazione di setup effettivamente utilizzata.

### <span style="color:blue"> Solido ($T^* = 0.8$, $\rho^* = 1.1$  ) </span>

Per compiere questa prima simulazione ho dovuto impostare $\mathcal{T} = 1.55$.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def read(filename) :
    file = open(filename,"r")
    value = []
    error = []

    
    for line in file:
        line = line.strip()
        lines = line.split()
        value.append(float(lines[2]))
        error.append(float(lines[3]))
        
    file.close()

    return value,error

def show_graphs (dirname,setup_temp) :

    value_press,error_press = read(dirname + "output_press.dat")
    value_temp,error_temp = read(dirname + "output_temp.dat")
    value_ekin,error_ekin = read(dirname + "output_ekin.dat")
    value_etot,error_etot = read(dirname + "output_etot.dat")
    value_epot,error_epot = read(dirname + "output_epot.dat")

    N=len(value_press)
    x=np.arange(N)

    plt.errorbar(x,value_press,yerr=error_press)
    plt.title("Pressione")
    plt.ylabel("Pressione")
    plt.xlabel("Blocchi considerati")
    plt.show()
    plt.errorbar(x,value_temp,yerr=error_temp)
    plt.title("Temperatura")
    plt.axhline(setup_temp,color="r")
    plt.ylabel("Temperatura")
    plt.xlabel("Blocchi considerati")
    plt.show()
    plt.errorbar(x,value_ekin,yerr=error_ekin)
    plt.title("Energia Cinetica per Particella")
    plt.ylabel("Energia Cinetica per Particella")
    plt.xlabel("Blocchi considerati")
    plt.show()
    plt.errorbar(x,value_epot,yerr=error_epot)
    plt.title("Energia Potenziale per Particella")
    plt.ylabel("Energia Potenziale per Particella")
    plt.xlabel("Blocchi considerati")
    plt.show()
    plt.errorbar(x,value_etot,yerr=error_etot)
    plt.title("Energia Totale per Particella")
    plt.ylabel("Energia Totale per Particella")
    plt.xlabel("Blocchi considerati")
    plt.show()
    
    print("Valori Finali \n Pressione :                         ",value_press[N-1]," +- ",error_press[N-1])
    print(" Temperatura :                       ",value_temp[N-1]," +- ",error_temp[N-1])
    print(" Energia Cinetica per Particella :   ",value_ekin[N-1]," +- ",error_ekin[N-1])
    print(" Energia Potenziale per Particella : ",value_epot[N-1]," +- ",error_epot[N-1])
    print(" Energia Totale per Particella :     ",value_etot[N-1]," +- ",error_etot[N-1])
    
def show_temp_graphs (dirname,setup_temp) :

    value_temp,error_temp = read(dirname + "output_temp.dat")
    
    N=len(value_temp)
    x=np.arange(N)

    plt.errorbar(x,value_temp,yerr=error_temp)
    plt.title("Temperatura")
    plt.ylabel("Temperatura")
    plt.axhline(setup_temp,color="r")
    plt.xlabel("Blocchi considerati")
    plt.show()

    print("Temperatura :                    ",value_temp[N-1]," +- ",error_temp[N-1])
    
show_temp_graphs("Solid_runs/Run_1.55_1.1/",0.8)
```

Ho ottenuto una convergenza molto rapida (sono stati sufficienti 4 blocchi da 5000 step ciascuno) per ottenere $\mathcal{T} \approx T^*$. Ho fatto un tentativo usando 20 blocchi (si può vedere qui sotto), ma ho notato che le differenze rispetto al tentativo con 4 blocchi sono trascurabili, quindi per comodità di calcolo ho utilizzato un numero ridotto di blocchi. /
Non sono riuscito in nessun caso ad ottenere in modo esatto una temperatura $T^* = 0.8$, ma ho ritenuto il valore ottenuto sufficientemente vicino da rendere trascurabili le differenze. Posso spiegare questa difficoltà ad ottenere una temperatura "esatta" con il fatto che i solidi tendono a formare strutture cristalline che risultano difficili da rompere. Sempre per questo motivo, ho dovuto scegliere $\mathcal{T} >> T^*$ in modo da avere energia sufficiente a rompere la struttura iniziale.

Vediamo ora qui sotto la prova con 20 blocchi e notiamo che non mostra differenze sostanziali con quella con 4 blocchi:

```python
show_temp_graphs("Solid_runs/Try_Setup_Run_1.542/",0.8)
```

Quindi, usata come simulazioni di setup quella con 4 blocchi, il risultato è:

```python
show_graphs("Solid_runs/FINAL_RUN/",0.8)
```

Anche nella simulazione finale non sono riuscito ad ottenere esattamente la $T^*$, ma ho ottenuto comunque un valore sufficientemente vicino da non presentare differenze significative. \
La materia ottenuta è chiaramente allo stato solido, infatti ha energia totale e pressione negative: questo fa sì che tenda a mantenere la struttura di partenza senza espandersi nello spazio circostante. L'energia data dalle forze di interazione interne domina chiaramente sull'energia cinetica delle particelle. 

Per quanto riguarda la convergenza dell'algoritmo, dopo alcune piccole oscillazioni iniziali notiamo che tutte le variabili di interesse rimangono fisse.

### <span style="color:blue"> Liquido ($T^* = 1.1$, $\rho^* = 0.8$  ) </span>

Anche in questo secondo caso ho scelto una simulazione di setup corta (4 blocchi). Sono tuttavia riuscito ad avvicinarmi più facilmente al valore $T^*$.

```python
show_temp_graphs("Liquid_runs/Run_1.95_0.8/",1.1)
```

Segue poi la simulazione completa:

```python
show_graphs("Liquid_runs/FINAL_RUN/",1.1)
```

Per quanto riguarda i valori ottenuti, capiamo di aver ottenuto un liquido dal fatto di avere un'energia totale negativa (anche se ovviamente maggiore rispetto a quella del caso solido) ad indicare la presenza importante di forze di interazione che mantengono costante il volume di suddetto liquido; ma abbiamo questa volta una pressione positiva, giustificata dalla proprietà dei liquidi di occupare il contenitore in cui sono posti.

### <span style="color:blue"> Gas ($T^* = 1.2$, $\rho^* = 0.05$  ) </span>

Per i gas, data la lentezza della convergenza ad un valore fissato di temperatura, ho deciso di fare una simulazione lunga per il setup, ho quindi usato 20 blocchi.

```python
show_temp_graphs("Gas_Runs/Run_0.95_0.05/",1.2)
```

Segue quindi la simulazione completa:

```python
show_graphs("Gas_Runs/FINAL_RUN/",1.2)
```

Notiamo immediatamente che nel caso del gas si è ottenuta facilmente la temperatura $T^*$ grazie ai legami estremamente deboli tra atomi che hanno permesso di trascurare più facilmente il fattore di interazione. \
Per quanto riguarda la termodinamica, notiamo che, coerentemente alle attese, l'energia totale e la pressione risultino entrambe positive, ad indicare il fatto che in un gas l'energia cinetica delle particelle domini chiaramente sulle forze interne.
